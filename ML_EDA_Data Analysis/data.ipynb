```python
data=pd.read_csv("dataset.csv")
```

```python
data=data.drop('Serial No.',axis=1)
```

```python
import matplotlib.pyplot as plt
def plot_scatter(Data,list1):
    for ri, rd in Data.iteritems():
        plt.scatter(rd,list1)
        plt.title(ri)
        plt.show()
```

```python
y=data['Result']
y
```

```python
plot_scatter(data,y)
```
